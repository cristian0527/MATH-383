#### P1; C
proportion of friends of that node that are female and the proportion of friends of that node that are at risk. Then compute the averages of these values across each group of students.

```python
female = []
risk = []

for node in G_8.nodes():
    count_f = 0
    count_r = 0
    for neighbor in G_8.neighbors(node):
        if G_8.nodes[neighbor]['Gender'] == 'FEMALE':
            count_f += 1
        if G_8.nodes[neighbor]['Risk'] == 1:
            count_r += 1
    female.append(count_f / float(G_8.degree(node)))
    risk.append(count_r / float(G_8.degree(node)))
```

```python
print(f"the proportion of friends of that node that are female: {np.mean(female)}")
print(f"the proportion of friends of that node that are at risk: {np.mean(risk)}")
```

#### P1; D

```python
from itertools import combinations
import pandas as pd

same_gender_edge, same_gender_no_edge = 0, 0
diff_gender_edge, diff_gender_no_edge = 0, 0

same_risk_edge, same_risk_no_edge = 0, 0
diff_risk_edge, diff_risk_no_edge = 0, 0

# ran these combinations since we are picking 2 nodes out of N -> G(8) nodes
for u, v in combinations(G_8.nodes(), 2):
    connected = G_8.has_edge(u, v)
    same_gender = G_8.nodes[u]['Gender'] == G_8.nodes[v]['Gender']
    same_risk = G_8.nodes[u]['Risk']   == G_8.nodes[v]['Risk']

    if same_gender and connected: same_gender_edge += 1
    if same_gender and not connected: same_gender_no_edge += 1
    if not same_gender and connected: diff_gender_edge += 1
    if not same_gender and not connected: diff_gender_no_edge += 1

    if same_risk and connected: same_risk_edge += 1
    if same_risk and not connected: same_risk_no_edge   += 1
    if not same_risk and connected: diff_risk_edge += 1
    if not same_risk and not connected: diff_risk_no_edge += 1
```

```python
# gender df
gender_df = pd.DataFrame(
    [[same_gender_edge, same_gender_no_edge],
     [diff_gender_edge, diff_gender_no_edge]],
    index = ['Same Gender', 'Different Gender'],
    columns = ['Edge', 'No Edge']
)

# risk df
risk_df = pd.DataFrame(
    [[same_risk_edge, same_risk_no_edge],
     [diff_risk_edge, diff_risk_no_edge]],
    index = ['Same Risk', 'Different Risk'],
    columns = ['Edge', 'No Edge']
)
```

#### P1; E

Create an additional contingency table where the entries count how many edges
observed in the network go between nodes with the same or different genders vs. how many edges in the network go between nodes with the same or different risk status. For example, an edge between two female students where one is at risk and one isn’t would go in the cell for same gender/different risk.

```python
same_gender_same_risk, same_gender_diff_risk = 0, 0
diff_gender_same_risk, diff_gender_diff_risk = 0, 0

for u, v in G_8.edges():
    same_gender = G_8.nodes[u]['Gender'] == G_8.nodes[v]['Gender']
    same_risk = G_8.nodes[u]['Risk'] == G_8.nodes[v]['Risk']

    if same_gender and same_risk: same_gender_same_risk += 1
    if same_gender and not same_risk: same_gender_diff_risk += 1
    if not same_gender and same_risk: diff_gender_same_risk += 1
    if not same_gender and not same_risk: diff_gender_diff_risk += 1
```

```python
gender_risk_df = pd.DataFrame(
    [[same_gender_same_risk, same_gender_diff_risk],
     [diff_gender_same_risk, diff_gender_diff_risk]],
    index   = ['Same Gender', 'Different Gender'],
    columns = ['Same Risk', 'Different Risk']
)
```

#### P2; A

```python
# container variable to help store the variables info
rows = []

for u, v in combinations(G_8.nodes(), 2):
    edge = int(G_8.has_edge(u, v))
    same_gender = int(G_8.nodes[u]['Gender'] == G_8.nodes[v]['Gender'])
    same_risk = int(G_8.nodes[u]['Risk'] == G_8.nodes[v]['Risk'])

    # number of shared friends
    shared_friends = len(list(nx.common_neighbors(G_8, u, v)))

    rows.append([u, v, edge, same_gender, same_risk, shared_friends])
```

```python
pairs_df = pd.DataFrame(rows, columns=['node i', 'node j', 'edge', 'same gender', 'same risk', 'shared friends'])

pairs_df.head()
```

#### P2; B

```python
# will be using scikit-learn for logistic models
from sklearn.linear_model import LogisticRegression

y = pairs_df['edge']

# logistic model 1 w/ same gender + same risk
X1 = pairs_df[['same gender', 'same risk']]
model1 = LogisticRegression(penalty=None, class_weight='balanced').fit(X1, y) # turn off penalty (regularization not applied)

# added this class weight = 'balanced' parameter because both models had the same accuracy score
# model was not learning --> tune the model to "study" the data

print("model 1 coefficients:")
print(f"intercept: {model1.intercept_[0]:.4f}")
print(f"same gender (B1): {model1.coef_[0][0]:.4f}")
print(f"same risk (B2): {model1.coef_[0][1]:.4f}")

# logistic model 2 w/ same gender + same risk + shared friends
X2 = pairs_df[['same gender', 'same risk', 'shared friends']]
model2 = LogisticRegression(penalty=None, class_weight='balanced').fit(X2, y) # turn off penalty (regularization not applied)

print("\nmodel 2 coefficients")
print(f"intercept: {model2.intercept_[0]:.4f}")
print(f"same gender (B1): {model2.coef_[0][0]:.4f}")
print(f"same risk (B2): {model2.coef_[0][1]:.4f}")
print(f"shared friends (B3): {model2.coef_[0][2]:.4f}")
```

#### P2; C

i wouldn't really say i'm too surprised, but it is interesting to get these results. our second model (logistic regression with the same gender, same risk, and same shared friends) performed better than the first model that had just taken into account the same gender and same risk as predictor variables. Since this jumped the accuracy by about 4%, we can say transitivity (shared friends) is an important feature to look at when we build our logistic regression model. --> later on build MCMC w/ specific thetas for these variables to score our models.

```python
# accuracy score gets this proportion
from sklearn.metrics import accuracy_score

prop1 = accuracy_score(y, model1.predict(X1))
prop2 = accuracy_score(y, model2.predict(X2))

print(f"model 1 Accuracy: {prop1:.4f}")
print(f"model 2 Accuracy: {prop2:.4f}")
```
