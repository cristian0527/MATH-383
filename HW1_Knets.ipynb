### Investigation will start here!

I want to investigate both graphs and see how they compare to each other using different centrality measures for the advice networks and friendship networks.

My goal is to ultimately see the overall viewpoint of who people go to for advice and who they are friends with between these two networks.

I plotted the graphs of both networks with all their edges to get a general idea of what both networks look like.

For the Advice (All) Network, since it is directed, the best centrality measure I thought to use would be degree, with this idea of how many nodes go to a specific individual. In order to do this, I used the in_degree function.

For the Friends (All) Network, I thought of using degree centrality and eigenvector centrality, since we would get a pretty good overview of the number of friends someone has, and how well connected a node is to important people.

After getting these centrality scores, I decided to rank them based on each score for their network, and then decided to make a dataframe to have a collective place to store the data and observe it.

There could have been different ways to rank the people, I was thinking about the possibility of finding a way to put weights on the different three rankings, but I just ended up doing a really simple way due to time. Probably also could have used the difference in friends' centrality rankings, too. I ended up just averaging the rankings between Advice and Friends networks and sorting them from low to high. This way, the lowest person with this average would be The first to show up, and thus I decided to rank them 1. Probably, not the most robust way to do it. But in the end, we got the ranking of importance (see code cell)! Also, when I was ranking, I did in degree ranking + eigenvector ranking. I did not check if there would be a difference if I used degree centrality ranking instead, but I would be interested in coming back to this and making this little switch.

For community identification, I decided to use modularity because of the density. Since in class we discussed of how greedy aggregation is fast, but can get stuck, I decided to use multi-level methods (Louvain/Leidan). With this, I got two different modularity scores for both graphs (0.112 for Advice, 0.113 for friends). Once I partitioned the nodes into communities, I was interested how often a node was in the same community for both networks, and we ended up with 1 out of 21 people (4.76%) for our simulated communities. Then I made side-by-side plots to see the different communities between both scenarios (Advice & Friends). Also, something to note, both networks had an output of 3 communities (though with extremely weak community structure in both cases). So, community assignments are essentially random between the two networks. I'm curious how different my results would be if I used the Spectral + refinement method. So based off this, there's not any meaningful communities. Though, I feel like I could have done something that is completely wrong.

In the future, I definitely would like to try different things:
1. Different method of ranking the 21 people
2. Comparing Louvain results with spectral clustering methods, which would be valuable
3. Using degree centrality instead of eigenvector centrality for friendship ranking
4. Measuring closeness centrality on top of the three and see what results are yield
5. Using one specific node perspective, rather than all get some insights to a specific person

```python
in_degree_advice = dict(advice_all.in_degree())
ranked_advice = sorted(in_degree_advice.items(), key=lambda x: x[1], reverse=True)
ranked_advice

print("In Degree Centrality")
for rank, (person, score) in enumerate(ranked_advice):
    print(f"{rank}. Person {person}: {score}")
```

```python
degree_friendship = dict(friends_all.degree())
ranked_degree = sorted(degree_friendship.items(), key=lambda x: x[1], reverse=True)
ranked_degree

eigenvector_friendship = nx.eigenvector_centrality(friends_all)
ranked_eigen = sorted(eigenvector_friendship.items(), key=lambda x: x[1], reverse=True)

print("Degree Centrality")
for rank, (person, score) in enumerate(ranked_degree):
    print(f"{rank}. Person {person}: {score}")

print("\n")

print("Eigenvector Centrality")
for rank, (person, score) in enumerate(ranked_eigen):
    print(f"{rank}. Person {person}: {score}")
```

### Since the network graphs are dense, I plan on using modularity for community detection!

```python
advice_undirected = advice_all.to_undirected()
louvain_advice = list(nx.algorithms.community.louvain_communities(advice_undirected))
for i, comm in enumerate(louvain_advice):
        print(f"  Community {i}: {sorted(list(comm))}")
```

```python
louvain_friend = list(nx.algorithms.community.louvain_communities(friends_all))
for i, comm in enumerate(louvain_friend):
        print(f"  Community {i}: {sorted(list(comm))}")
```

```python
from networkx.algorithms.community import modularity

modularity_advice = modularity(advice_undirected, louvain_advice)
modularity_friend = modularity(friends_all, louvain_friend)
```

```python
louvain_advice_dict = dict()
for i in range(len(louvain_advice)):
    for node in list(louvain_advice[i]):
        louvain_advice_dict[node] = i

louvain_friend_dict = dict()
for i in range(len(louvain_friend)):
    for node in list(louvain_friend[i]):
        louvain_friend_dict[node] = i
```

```python
comparison_data = []
for person in range(21):
    comparison_data.append({
        'Person': person,
        'Advice_Community': louvain_advice_dict[person],
        'Friend_Community': louvain_friend_dict[person],
        'Same_Community': louvain_advice_dict[person] == louvain_friend_dict[person]
    })

knets_communities = pd.DataFrame(comparison_data)
knets_communities['Person'] = knets_communities['Person'] + 1
knets_communities['Friend_Community'] = knets_communities['Friend_Community'] + 1
knets_communities['Advice_Community'] = knets_communities['Advice_Community'] + 1
knets_communities.set_index('Person', inplace=True)

knets_communities
```

```python
same_community = knets_communities['Same_Community'].sum()
proportion_same = same_community / len(knets_communities)
print(f"\n{same_community} out of 21 people ({proportion_same:.2%}) are in matching communities")
```
